```python
def GetTitle(name):
    title_group = {'mr': 'Mr', 'mrs': 'Mrs', 'miss': 'Miss', 'master': 'Master', 'don': 'Sir', 'rev': 'Sir', 'dr': 'Officer', 'mme': 'Mrs', 'ms': 'Mrs',
       'major': 'Officer', 'lady': 'Lady', 'sir': 'Sir', 'mlle': 'Miss', 'col': 'Officer', 'capt': 'Officer', 'the countess': 'Lady',
       'jonkheer': 'Sir', 'dona': 'Lady'}    
    first_name_title = name.split(',')[1]
    title = first_name_title.split('.')[0]
    title = title.strip().lower()
    return title_group[title]
```

```python
df['Title'] = df.Name.map(lambda x: GetTitle(x))
```

```python
df.head()
```

```python
df[df.Age.notnull()].boxplot('Age', 'Title');
```

```python
title_age_median= df.groupby('Title').Age.transform('median')
df.Age.fillna(title_age_median, inplace=True)
```
